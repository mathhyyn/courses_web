```python
import datetime


def gift_count(budget, month, birthdays):
    bday_staff = []
    for name, birthday in birthdays.items():
        if birthday.month == month:
            bday_staff.append((name, birthday))
    staff_strs = []
    for name, birthday in sorted(bday_staff, key=lambda x: x[1].day):
        staff_strs.append(f"{name} ({birthday.strftime('%d.%m.%Y')})")
    num = len(bday_staff)
    if num == 0:
        print("В этом месяце нет именинников.")
    else:
        print(
            f"Именинники в месяце {month}: {', '.join(staff_strs)}. При бюджете {budget} они получат по {budget // num} рублей."
        )


birthdays = {
    "Иванов Иван Иванович": datetime.date(1989, 5, 1),
    "Петров Петр Петрович": datetime.date(1998, 5, 6),
}
gift_count(20000, 5, birthdays)
gift_count(budget=20001, month=5, birthdays=birthdays)
```

```python
def lists_sum(*args, unique=False):
    total_sum = 0
    merged_list = []
    for arg in args:
        merged_list.extend(arg)
    if unique:
        total_sum = sum(set(merged_list))
    else:
        total_sum = sum(merged_list)
    return total_sum


assert lists_sum([1, 1], [1], [1, 2, 3]) == 9
assert lists_sum([1, 1, 1], [1, 1], unique=True) == 1
assert lists_sum([1, 1, 1], unique=False) == 3
```

```python
def get_balance(name, transactions):
    balance = 0
    for transaction in transactions:
        if transaction["name"] == name:
            balance += transaction["amount"]

    return balance


def count_debts(names, amount, transactions):
    debts = {}
    for name in names:
        balance = get_balance(name, transactions)
        debts[name] = max(amount - balance, 0)

    return debts


transactions = [
    {"name": "Василий", "amount": 500},
    {"name": "Петя", "amount": 100},
    {"name": "Василий", "amount": -300},
]

assert get_balance("Василий", transactions) == 200
assert count_debts(["Василий", "Петя", "Вова"], 150, transactions) == {
    "Василий": 0,
    "Петя": 50,
    "Вова": 150,
}
```

```python
def check_string(string):
    import re

    phone_regex = r"(((\+7|8|7)-?|)\d\d\d-?\d\d\d-?\d\d-?\d\d)|((\+7|8|7|) ?\(\d\d\d\) ?\d\d\d-?\d\d-?\d\d)"
    email_regex = r"[A-Za-z0-9._+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
    phone_match = re.fullmatch(phone_regex, string)
    email_match = re.fullmatch(email_regex, string)

    if email_match or phone_match:
        return True
    return False


assert check_string("+7-916-000-00-00")
assert check_string("+7 (916) 000-00-00")
assert check_string("abc@abc.ab.ab")
```

```python
def get_popular_name_from_file(filename):
    names = {}

    with open(filename) as file:
        for line in file:
            first_name, _ = line.strip().split(" ")
            if not first_name in names:
                names[first_name] = 0
            names[first_name] += 1

    max_count = max(names.values())
    popular_names = [name for name, count in names.items() if count == max_count]
    popular_names.sort()
    return ", ".join(popular_names)


filename = "names.txt"
popular_name = get_popular_name_from_file(filename)
print(popular_name)
```

```python
class Calculator:
    last = None

    def __init__(self):
        self.operations = []

    @classmethod
    def clear(cls):
        cls.last = None

    def history(self, n):
        return None if len(self.operations) < n else self.operations[n - 1]

    def updateHistory(self, method, a, b, result):
        str_history = f"{method}({a}, {b}) == {round(result, 1)}"
        Calculator.last = str_history
        self.operations.insert(0, str_history)

    def sum(self, a, b):
        result = a + b
        self.updateHistory("sum", a, b, result)
        return result

    def sub(self, a, b):
        result = a - b
        self.updateHistory("sub", a, b, result)
        return result

    def mul(self, a, b):
        result = a * b
        self.updateHistory("mul", a, b, result)
        return result

    def div(self, a, b, mod=False):
        result = a % b if mod else a / b
        self.updateHistory("div", a, b, result)
        return result


calc = Calculator()
result = calc.sum(5, 15)
print(calc.history(1))
print(Calculator.last)
```

```python
class BaseWallet:
    def __init__(self, name, amount, exchange_rate):
        self.name = name
        self.amount = amount
        self.exchange_rate = exchange_rate

    def to_base(self):
        return self.amount * self.exchange_rate

    def from_base(self, amount):
        return amount / self.exchange_rate

    def spend_all(self):
        self.amount = min(self.amount, 0)

    def __add__(self, wallet2):
        if isinstance(wallet2, (int, float)):
            return self.__class__(self.name, self.amount + wallet2)

        if isinstance(wallet2, BaseWallet):
            result_amount = self.to_base() + wallet2.to_base()
            return self.__class__(self.name, self.from_base(result_amount))

    def __radd__(self, wallet2):
        if isinstance(wallet2, (int, float)):
            return self.__class__(self.name, self.amount + wallet2)

    def __sub__(self, wallet2):
        if isinstance(wallet2, BaseWallet):
            result_amount = self.to_base() - wallet2.to_base()
            return self.__class__(self.name, self.from_base(result_amount))

        if isinstance(wallet2, (int, float)):
            result_amount = self.amount - wallet2
            return self.__class__(self.name, result_amount)

    def __rsub__(self, wallet2):
        if isinstance(wallet2, (int, float)):
            return self.__class__(self.name, wallet2 - self.amount)

    def __mul__(self, num):
        return self.__class__(self.name, self.amount * num)

    def __rmul__(self, num):
        return self.__class__(self.name, self.amount * num)

    def __truediv__(self, num):
        return self.__class__(self.name, self.amount / num)

    def __rtruediv__(self, num):
        return self.__class__(self.name, self.amount / num)

    def __eq__(self, wallet2):
        return self.__class__ == wallet2.__class__ and self.amount == wallet2.amount

    def __str__(self):
        return f"{self.class_name} {self.name} {self.amount}"


class RubbleWallet(BaseWallet):
    def __init__(self, name, amount):
        self.class_name = "Rubble Wallet"
        super().__init__(name, amount, 1)


class DollarWallet(BaseWallet):
    def __init__(self, name, amount):
        self.class_name = "Dollar Wallet"
        super().__init__(name, amount, 60)


class EuroWallet(BaseWallet):
    def __init__(self, name, amount):
        self.class_name = "Euro Wallet"
        super().__init__(name, amount, 70)


ruble_wallet = RubbleWallet("Ruble", 5000)
dollar_wallet = DollarWallet("Dollar", 100)
euro_wallet = EuroWallet("Euro", 200)

print(2 * (200 - dollar_wallet + 5) * 2)

ruble_wallet /= 5
assert ruble_wallet == RubbleWallet("A", 1000)

wallet1 = RubbleWallet("X", 100)
result1 = wallet1 / 2
print(result1)
```

```python
import re


class Field:
    def __init__(self):
        self.data = {}

    def getByLetterDigit(self, letter, digit):
        if not (letter.isalpha() and len(letter) == 1):
            raise ValueError(f"Invalid letter: {letter}")
        return f"{letter.upper()}{digit}"

    def createKey(self, key):
        if isinstance(key, tuple):
            if len(key) != 2:
                raise TypeError(f"Invalid key: {key}")
            left, right = key
            if isinstance(left, float) or isinstance(right, float):
                raise ValueError(f"Invalid key: {key}")
            if isinstance(left, int) and isinstance(right, str):
                return self.getByLetterDigit(right, left)
            elif isinstance(left, str) and isinstance(right, int):
                return self.getByLetterDigit(left, right)
            elif isinstance(left, str) and isinstance(right, str):
                if left.isdigit():
                    return self.getByLetterDigit(right, int(left))
                elif right.isdigit():
                    return self.getByLetterDigit(left, int(right))
                else:
                    raise ValueError(f"Invalid key: {key}")
            else:
                raise TypeError(f"Invalid key type: {key}")
        elif isinstance(key, str):
            match = re.fullmatch(r"([A-Z])([1-9]\d*)|([1-9]\d*)([A-Z])", key.upper())
            if not match:
                raise ValueError(f"Invalid key: {key}")
            if match[1]:  # format A1
                return self.getByLetterDigit(match[1], match[2])
            if match[3]:  # format 1A
                return self.getByLetterDigit(match[4], match[3])
        else:
            raise TypeError(f"Invalid key type: {type(key)}")

    def __getitem__(self, key):
        return self.data.get(self.createKey(key))

    def __setitem__(self, key, value):
        self.data[self.createKey(key)] = value

    def __delitem__(self, key):
        del self.data[self.createKey(key)]

    def __contains__(self, key):
        return self.createKey(key) in self.data

    def __iter__(self):
        return iter(self.data.values())


field = Field()

field[1, "a"] = 25
field["a", 1] = 25
field["a", "1"] = 25
field["1", "a"] = 25
field["1a"] = 25
field["a1"] = 25
field[1, "A"] = 25
field["A", 1] = 25
field["A", "1"] = 25
field["1", "A"] = 25
field["1A"] = 25
field["A1"] = 25

print(field[1, "a"])  # 25
print(field["a", 1])  # 25
print(field["a", "1"])  # 25
print(field["1", "a"])  # 25
print(field["1a"])  # 25
print(field["a1"])  # 25
print(field[1, "A"])  # 25
print(field["A1"])  # 25

del field["A1"]
print("A1" in field)  # False
field["A2"] = 26
field["B1"] = 27

for value in field:
    print(value)
```

```python
from datetime import datetime
import json
from api import register_booking


def create_booking(room_name, start, end) -> str:
    print("Начинаем создание бронирования")

    result = False
    msg = "Бронирование не начато"

    try:
        booking = Booking(room_name, start, end)
        result = register_booking(booking)

        if result:
            msg = "Бронирование создано"
        else:
            msg = "Комната занята"

    except KeyError:
        result = False
        msg = "Комната не найдена"

    except ValueError:
        result = False
        msg = "Ошибка: неверные даты бронирования"

    finally:
        print("Заканчиваем создание бронирования")

    booking_data = {
        "created": result,
        "msg": msg,
        "booking": {
            "room_name": booking.room_name,
            "duration": booking.duration,
            "start_date": booking.start_date,
            "end_date": booking.end_date,
            "start_time": booking.start_time,
            "end_time": booking.end_time,
        },
    }
    return json.dumps(booking_data)


class Booking:
    def __init__(self, room_name, start_datetime, end_datetime):
        if end_datetime < start_datetime:
            raise ValueError("End datetime should be greater than start datetime")

        self.room_name = room_name
        self.start = start_datetime
        self.end = end_datetime

    @property
    def duration(self):
        return int((self.end - self.start).total_seconds() // 60)

    @property
    def start_date(self):
        return self.start.strftime("%Y-%m-%d")

    @property
    def end_date(self):
        return self.end.strftime("%Y-%m-%d")

    @property
    def start_time(self):
        return self.start.strftime("%H:%M")

    @property
    def end_time(self):
        return self.end.strftime("%H:%M")
```

```python
def logging_decorator(logger):
    def decorator(func):
        import time
        from datetime import datetime
        from inspect import getcallargs

        def wrapper(*args, **kwargs):
            call_time = datetime.now()
            result = func(*args, **kwargs)
            log_entry = {
                "name": func.__name__,
                "arguments": getcallargs(func, *args, **kwargs),
                "call_time": call_time,
                "result": result,
            }
            logger.append(log_entry)
            return result

        return wrapper

    return decorator


logger = []  # этот словарь будет хранить наш "лог"


@logging_decorator(logger)  # в аргументы фабрики декораторов подается логгер
def test_simple(a, b=2):
    return 127


# при вызове функции в список logger должен добавиться словарь с
test_simple(1)
# информацией о вызове функции

print(logger)
```
